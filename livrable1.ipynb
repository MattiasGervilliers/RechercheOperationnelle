<h2>Reformulation du problème</h2>
<p>
À la suite de l’appel à projets lancé par l’ADEME, notre structure est missionnée pour répondre à un problème algorithmique relevant du domaine de la mobilité intelligente. Ce problème consiste à <strong>déterminer, sur un réseau routier, une tournée optimale permettant de relier entre elles un sous-ensemble de villes, puis de revenir au point de départ, tout en minimisant la durée totale du trajet</strong>.
</p><p>
Il s’agit donc d’un problème d’optimisation, c’est-à-dire d’un problème où l’on cherche à minimiser ou maximiser une fonction définie sur un ensemble de solutions possibles. Dans notre cas, l’objectif est de minimiser la fonction représentant la durée totale de la tournée, sur l’ensemble des itinéraires réalisables reliant ces villes.
</p><p>
Deux approches théoriques s’offrent à nous pour traiter ce type de problème :
</p><p>
L’approche analytique, qui vise à obtenir une solution exacte en utilisant des outils mathématiques (calcul différentiel, systèmes d’équations, etc.). Cette méthode est rigoureuse mais devient rapidement inapplicable pour des problèmes de grande taille ou présentant une forte complexité combinatoire.
</p><p>
L’approche numérique, qui repose sur l’utilisation d’algorithmes informatiques pour approximer une solution optimale. Bien que le résultat soit approché, cette méthode est indispensable lorsque la résolution analytique est inaccessible ou trop coûteuse en pratique.
</p>

<h3>Les données</h3>
<h4><i>Modélisation</i></h4>
<p>
Nous devons représenter des données qui décrivent un réseau urbain, soit : des villes, et des routes reliant ces villes. Le meilleur moyen de représenter ce cas est par un graphe pondéré non orienté. En effet, chaque ville peut être représenté par un sommet, chaque route par une arrête, chaque distance entre chaque ville peut être renseignée comme le poids assigné à chaque arrête.
</p><p>
Ainsi, les données sont modélisées par un graphe pondéré G=(V,E,w), où :
<li>V est l’ensemble des sommets (villes),</li>
<li>E⊆V×V est l’ensemble des arêtes (routes),</li>
<li>w:E→R+ est une fonction de poids représentant la durée ou distance associée à chaque arête.</li>

<h3>Le problème</h3>
<p>
Nous avons vaguement énoncé le problème précédemment, mais concentrons-nous maintenant sur sa réelle représentation, par rapport à un graphe.
</p><p>
Il faut noter que quelques notions sont à définir, pour pouvoir expliquer plus clairement. Effectivement un chemin qui revient à son point de départ dans un graphe est appelé cycle, et le fait de passer une seule fois par chaque sommet est appelé chemin hamiltonien.
</p><p>
Notre problème se rapproche donc d’un cycle hamiltonien, à la différence que nous ne voulons pas forcément passer une seule fois par chaque sommet, mais au moins une fois.
</p>

Programme linéaire
Dans la recherche du programme linéaire lié à notre problème, nous allons passer par plusieurs étapes.
Variable de décision
Identifions les inconnues sachant que nous allons passer par v arrêtes, n fois, en voulant minimiser la durée k, somme du poids des arrêtes.
Donc :
	Soit G le graphe dans lequel on évolue.
	Soit v le nombre d’arrête que l’on traverser.
	Soit n le nombre de fois où l’on passer par cette arrête.
	Soit k le poids d’une arrête.

Les contraintes sont des conditions sur l'usage des ressources, c'est à dire des conditions que doivent respecter les valeurs des variables de décision pour former une solution admissible.

Fonction économique
La fonction économique décrit ce qu’on veut optimiser lors de l’usage de nos ressources. C’est donc une fonction de nos variables de décision, que l’on doit maximiser ou minimiser. Comme pour les contraintes, dans un programme linéaire, cette fonction économique doit être elle-même linéaire.
